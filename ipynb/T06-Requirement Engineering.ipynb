<br>‚ö† <b>Tip</b><br>
>Requirements language from the stakeholders often includes ‚Äòmust‚Äô, ‚Äòshould‚Äô, etc phrases, and these may indeed be their priorities.
The analyst however during the work of refining the requirements should write the requirements in a way that doesn‚Äôt imply a priority in the requirement‚Äôs expression itself.
Thus for example requirement 2 which states ‚ÄòThe system must record customer orders ...‚Äô should be re-phrased as ‚ÄòThe system will or shall record customer orders ...‚Äô and then assign ‚Äòmust‚Äô as the priority.
